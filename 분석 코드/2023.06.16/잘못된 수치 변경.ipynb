```python
index = ['대기업','중견기업','중소기업','자동차','전기/전자','헬스케어','조선', '유통물류','철강','섬유화학','에너지','기계','소재/뿌리']


## 인식도 전환전략 역량지수 

data_worng = [[2.37,2.39,2.11,2.45,2.31,2.12,1.97,2.09,2.18,2.23,1.91,2.33,1.92],
             [1.99,1.73,1.51,1.54,1.82,1.60,1.50,1.41,1.39,1.64,1.38,1.86,1.40],
             [2.21,1.77,1.62,1.61,2.02,1.73,1.59,1.45,1.41,1.78,1.44,2.01,1.53]]
data_fact = [[2.77,2.75,2.28,2.83,2.62,2.25,2.06,2.27,2.47,2.45,1.93,2.62,2.02],
            [2.10,1.83,1.55,1.65,1.91,1.63,1.52,1.46,1.47,1.70,1.38,1.94,1.42],
            [2.26,1.82,1.64,1.66,2.06,1.74,1.60,1.47,1.45,1.81,1.44,2.05,1.55]]



```

## 기업별

```python
plt.figure(figsize=(12,8))

plt.subplot(311)
plt.plot(index[:3], data_worng[0][:3] ,color = 'black',  label = '수정전')
plt.plot(index[:3], data_fact[0][:3] ,color = 'red',  label = '수정후')
xlabel=index[:3]
plt.title='인식도'
plt.legend()


plt.subplot(312)
plt.plot(index[:3], data_worng[1][:3] ,color = 'black',  label = '수정전')
plt.plot(index[:3], data_fact[1][:3] ,color = 'red',  label = '수정후')
plt.title='디지털전환 전략'
xlabel=index[:3]
plt.legend()


plt.subplot(313)
plt.plot(index[:3], data_worng[2][:3] ,color = 'black',  label = '수정전')
plt.plot(index[:3], data_fact[2][:3] ,color = 'red',  label = '수정후')
xlabel=index[:3]
plt.title='산업디지털전환 역량지수'
plt.legend()

```

## 산업별

```python
plt.figure(figsize=(12,8))

plt.subplot(311)
plt.plot(index[3:], data_worng[0][3:] ,color = 'black',  label = '수정전')
plt.plot(index[3:], data_fact[0][3:] ,color = 'red',  label = '수정후')
xlabel=index[3:]
plt.title='인식도'
plt.legend()


plt.subplot(312)
plt.plot(index[3:], data_worng[1][3:] ,color = 'black',  label = '수정전')
plt.plot(index[3:], data_fact[1][3:] ,color = 'red',  label = '수정후')
xlabel=index[3:]
plt.title='디지털전환 전략'
plt.legend()


plt.subplot(313)
plt.plot(index[3:], data_worng[2][3:] ,color = 'black',  label = '수정전')
plt.plot(index[3:], data_fact[2][3:] ,color = 'red',  label = '수정후')
xlabel=index[3:]
plt.title='산업디지털전환 역량지수'
plt.legend()
```

```python
data_1 = pd.DataFrame({'분류':index , 
                   '수정전':data_worng[0],
                   '수정후':data_fact[0]})

data_2 = pd.DataFrame({'분류':index , 
                   '수정전':data_worng[1],
                   '수정후':data_fact[1]})

data_3 = pd.DataFrame({'분류':index , 
                   '수정전':data_worng[2],
                   '수정후':data_fact[2]})
```

```python
data_3
```

```python
data_11=data_1.loc[:2]
data_12=data_1.loc[3:]

data_21=data_2.loc[:2]
data_22=data_2.loc[3:]

data_31=data_3.loc[:2]
data_32=data_3.loc[3:]
```

```python
data_11['수정전 순위'] = data_11['수정전'].rank(method='first', ascending=False)
data_11['수정후 순위'] = data_11['수정후'].rank(method='first', ascending=False)
data_12['수정전 순위'] = data_12['수정전'].rank(method='first', ascending=False)
data_12['수정후 순위'] = data_12['수정후'].rank(method='first', ascending=False)


data_21['수정전 순위'] = data_21['수정전'].rank(method='first', ascending=False)
data_21['수정후 순위'] = data_21['수정후'].rank(method='first', ascending=False)
data_22['수정전 순위'] = data_22['수정전'].rank(method='first', ascending=False)
data_22['수정후 순위'] = data_22['수정후'].rank(method='first', ascending=False)


data_31['수정전 순위'] = data_31['수정전'].rank(method='first', ascending=False)
data_31['수정후 순위'] = data_31['수정후'].rank(method='first', ascending=False)
data_32['수정전 순위'] = data_32['수정전'].rank(method='first', ascending=False)
data_32['수정후 순위'] = data_32['수정후'].rank(method='first', ascending=False)
```

```python
data_111 = data_11.sort_values(by='수정전 순위',axis=0)
data_112 = data_11.sort_values(by='수정후 순위',axis=0)

data_211 = data_21.sort_values(by='수정전 순위',axis=0)
data_212 = data_21.sort_values(by='수정후 순위',axis=0)

data_311 = data_31.sort_values(by='수정전 순위',axis=0)
data_312 = data_31.sort_values(by='수정후 순위',axis=0)
```

```python
data_121 = data_12.sort_values(by='수정전 순위',axis=0)
data_122 = data_12.sort_values(by='수정후 순위',axis=0)

data_221 = data_22.sort_values(by='수정전 순위',axis=0)
data_222 = data_22.sort_values(by='수정후 순위',axis=0)

data_321 = data_32.sort_values(by='수정전 순위',axis=0)
data_322 = data_32.sort_values(by='수정후 순위',axis=0)
```

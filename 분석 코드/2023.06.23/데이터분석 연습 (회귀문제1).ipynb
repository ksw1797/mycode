```python
#### 데이터 불러오기 

train = pd.read_csv('C:/Users/user/Desktop/Python/train.csv')
test = pd.read_csv('C:/Users/user/Desktop/Python/test.csv')
submission = pd.read_csv('C:/Users/user/Desktop/Python/sample_submission.csv')


### 데이터 크기, 분포확인
# print(train.shape , test.shape , submission.shape)
# print(train.corr())
# print(train.isna().sum() , test.isna().sum())

target = train['price']
train.drop('price',axis = 1 , inplace = True)

#test.drop('model',axis=1,inplace=True)
# print(train.describe(include = 'object'))
#print(train.describe(exclude = 'object'))


## 인사이트 모델을 빼고 mileage 는 로그 변환 혹은 스케일링을 해주면 좋을듯 (이상치) 모델은 빼는것이 좋아보임 너무많음 일단 돌려보고

## 훈련세트 스케일링 두가지로 하고 비교
import sklearn 
import sklearn.preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

#scale_minmax = MinMaxScaler()
#train[['year', 'mileage', 'tax','mpg','engineSize']] = scale_minmax.fit_transform(train[['year', 'mileage', 'tax','mpg','engineSize']])
#test[['year', 'mileage', 'tax','mpg','engineSize']] = scale_minmax.transform(test[['year', 'mileage', 'tax','mpg','engineSize']])

#scale_std = StandardScaler()
#train[['year', 'mileage', 'tax','mpg','engineSize']] = scale_std.fit_transform(train[['year', 'mileage', 'tax','mpg','engineSize']])
#test[['year', 'mileage', 'tax','mpg','engineSize']] = scale_std.transform(test[['year', 'mileage', 'tax','mpg','engineSize']])


#print(train.describe(exclude = 'object'))
train= pd.get_dummies(train)
test = pd.get_dummies(test)

from sklearn.model_selection import train_test_split

X_train, X_val , Y_train , Y_val = train_test_split(train , target , test_size = 0.2 , random_state = 123)

#print(X_train.shape,X_val.shape,Y_train.shape,Y_val.shape)


import sklearn.ensemble 
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()
model_rf.fit(X_train,Y_train)
pred = model_rf.predict(X_val)

#### 하이퍼 파라미터 최적화 작업 및 학습 재시도

# print(sklearn.model_selection.__all__)
from sklearn.model_selection import GridSearchCV

#par = { 'max_depth' : [13,14,15,16,17],'min_samples_split' : [2, 3]}


#grid_dt  = GridSearchCV(estimator=model_rf, param_grid=par, cv=5, refit=True)
#grid = grid_dt.fit(X_train,Y_train)

#print(grid.best_params_) # {'max_depth': 15, 'min_samples_split': 3}
#print(grid.best_score_) # 0.9179437904756202

## 학습 재시도

model_rf2 = RandomForestRegressor(max_depth= 14, min_samples_split= 3,random_state = 123)
model_rf2.fit(X_train,Y_train)
pred2 = model_rf2.predict(X_val)

### 평가 지표  RMSE
import sklearn.metrics
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(Y_val,pred) ** 0.5
rmse2 = mean_squared_error(Y_val,pred2) ** 0.5

print(rmse,rmse2) 
########## 테스트셋 학습

result = model_rf.predict(test)
result2 = model_rf2.predict(test)
submission['price'] = result2
submission.to_csv('submission.csv',index = False)

```
